### **How List are store in Memory?**

**1 The list object itself is stored in memory**

- Lets say the list starts at address 1000.

- Each slot of the list (like an array cell) holds a reference (address) to the actual object.

**2 The actual objects (1, 2, 3, 4) are stored separately**

- They live elsewhere in memory at addresses like 5000, 6000, 

- At those addresses, the binary representation of the numbers is stored

image.png

```python
[5000,6000, 7000, 8000]
```

 **Step-by-Step Explanation**

**List container:**

Lives at some memory region (10001003).

Each slot stores a pointer (reference), not the actual value.

**Pointers inside list:**

Example: At index 0 (address 1000), Python stores 5000.

5000 is the memory address where the object 1 actually lives.

**Actual object storage:**

At address 5000, binary 00000001 is stored.

That binary means  integer value 1.

**So retrieval works like this:**

You ask: my_list[0]  Python looks at slot 1000  finds 5000  goes to address 5000  reads binary 00000001  returns 1.

**Repeat for other indexes.**

** In Real Life Terms:**

Think of the list as a contact list in your phone:

At position 0  you dont store your friends face, you store their phone number (address).

The actual friend (value) is somewhere else in the world (memory).

To meet them, you use the phone number (pointer) to reach them.

### **Length of an Object in Python**

In Python, everything is an object (strings, lists, dictionaries, user-defined classes, etc.).

If an object has a length (like a string or list), it means it contains multiple elements inside it.

The built-in function len(object) is used to find how many items the object contains.

But not every object has a length  only objects that define the special method __len__().

```python
s = "Hello"
print(len(s))   # 5  because string has 5 characters
```

```python
L = [10, 20, 30, 40]
print(len(L))   # 4  because 4 items are inside
```

Concept is the same for all objects  length = number of contained items.

Difference is the container type:

- For a string, length = number of characters.
- For a list, length = number of elements.
- For a dictionary, length = number of key-value pairs.
- For a set, length = number of unique elements.

```python
print(len("Python"))            # 6 (characters)
print(len([1, 2, 3, 4]))        # 4 (list elements)
print(len({"a": 1, "b": 2}))    # 2 (dictionary items)
print(len({1, 2, 3, 4, 4}))     # 4 (set has unique items)
```

### **Slicing**

list[start : stop : step]

- start = index where slicing begins (inclusive).
- stop = index where slicing ends (exclusive).
- step = interval (positive  forward, negative  backward).


If len(L) = n,

L[0 : n]  gives the whole list.

Because stop = n (one beyond last index n-1).

 Think of stop as the "boundary line" after the last index you want.

```python
# Slicing
L = [1,2,3,4,5,6]
# length of list is 6, and index is 0 to 5
print(L[0:5])
# print element from 0 index to 5 index
print(L[0:6])
print(len(L))
# Directly ask for specific element by using index
print(L[5])
#print(L[6])
```

```python
print(len(L))
print(len(L)-1)
```

### **Why Slicing Doesnt Raise Index Errors**

```python
L = [1,2,3,4]
print(L[0:9])
```

**Why no error?**

When you do indexing like L[9], Python checks if index 9 exists. Since len(L)=4 (last valid index = 3), it throws IndexError.

But in slicing, Python is smart:

- Start = 0  valid

- Stop = 9  it sees this is beyond the list length, so it just safely stops at the end of the list.

No error, just returns all available elements.

** Rule of thumb:**

**Indexing** = strict, must be within 0  len-1.

**Slicing** = safe, stop value can go beyond len, Python will just stop at the end.

# Zip
The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# iterator = list
```

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```

```python
# by Loops
L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]
for i in range(L1[i]):
  for i in range(L2[i]):
    print(L1[i]+L2[i])
```
